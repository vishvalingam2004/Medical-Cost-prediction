```python
# Preprocessing
from sklearn.preprocessing import OrdinalEncoder
```

```python
enc=OrdinalEncoder()
```

```python
a=['sex']
b=['smoker']
c=['region']
ds[a]=enc.fit_transform(ds[a])
ds[b]=enc.fit_transform(ds[b])
ds[c]=enc.fit_transform(ds[c])
```

```python
ds
```

```python
ds=ds.astype({"sex":'int',"smoker":'int',"region":'int'})
```

```python
# Train-Test Building.
from sklearn.model_selection import train_test_split
```

```python
x=ds.iloc[:,0:6]
y=ds["charges"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
x_train
```

```python
# Machine Learning Model Building.
from sklearn.linear_model import LinearRegression
```

```python
mod=LinearRegression()
```

```python
mod.fit(x_train,y_train)
```

```python
pred=mod.predict(x_test)
print(pred)
```

```python
# Calculating R Squared Value.
from sklearn.metrics import r2_score
```

```python
val=r2_score(y_test,pred)
print("R Squared Value is:",val*100)
```

```python
sns.regplot(data=ds,x=y_test,y=pred,color="blue")
```
